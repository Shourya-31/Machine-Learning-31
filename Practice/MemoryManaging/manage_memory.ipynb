```python
## Handling Circular reference
import gc

class MyObject:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created")

    def __del__(self):
        print(f"Object {self.name} deleted")

# Create circular reference
obj1 = MyObject("obj1")
obj2 = MyObject("obj2")
obj1.ref = obj2
obj2.ref = obj1

del obj1
del obj2

## Manually trigger the garbage collection
gc.collect()
```

```python
print(f"Garbage Collected Objects:{gc.garbage}")
```

```python
## Genrators for memory efficiency
# Allows only 1 item production ata  time

def generate_num(n):
    for i in range(n):
        yield i

# using generator
for num in generate_num(100000):
    print(num)
    if num>10:
     break
```

```python
## Profiling Memory usage with Tracemalloc
import tracemalloc   # Import the tracemalloc module to track memory allocations in Python

def create_list():
    # Function that creates and returns a list containing numbers from 0 to 9999
    return [i for i in range(10000)]

def main():
    # This block runs only if the script is executed directly (not imported as a module)
    if __name__ == "__main__":
        tracemalloc.start()  # Start tracing memory allocations

        create_list()  # Call the function that allocates memory by creating a list

        # Take a snapshot of current memory usage
        snapshot = tracemalloc.take_snapshot()

        # Get memory allocation statistics, grouped by the line number in the source code
        top_stats = snapshot.statistics('lineno')

        print("[Top 10]")  # Display header for the memory usage report

        # Loop through and print each memory statistic
        # [::] means all elements; typically you'd slice to top_stats[:10] for top 10 entries
        for stat in top_stats[::]:
            print(stat)
```

```python
main()
```
