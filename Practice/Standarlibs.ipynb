```python
# Importing the built-in `csv` module, which provides functionality to read from and write to CSV (Comma-Separated Values) files.
import csv 

# Open a file named 'example.csv' in write mode ('w').
# `newline=''` ensures that no extra blank lines are added on Windows systems.
with open('example.csv', mode='w', newline='') as file:
    
    # Create a CSV writer object that will write to the opened file.
    writer = csv.writer(file)
    
    # Write the header row (column names) to the CSV file.
    writer.writerow(['name', 'age'])
    
    # Write a second row of data to the CSV file.
    writer.writerow(['Sho', '23'])

# Open the same 'example.csv' file, this time in read mode ('r').
with open('example.csv', mode='r') as file:
    
    # Create a CSV reader object to read from the file.
    reader = csv.reader(file)
    
    # Loop through each row in the CSV file and print it.
    for row in reader:
        print(row)
        
```

```python
# Importing `datetime` and `timedelta` classes from the `datetime` module.
# `datetime` is used to work with dates and times.
# `timedelta` is used to represent a duration (difference between two dates or times).
from datetime import datetime, timedelta

# Get the current local date and time using `datetime.now()`.
# It returns a `datetime` object representing the current moment.
now = datetime.now()

# Print the current date and time.
print(now)

# Subtract 1 day from the current time using `timedelta(days=1)`.
# This gives the date and time exactly 24 hours before `now` (i.e., yesterday).
yesterday = now - timedelta(days=1)

# Print the date and time for yesterday.
print(yesterday)
```

```python
# Importing the `time` module, which provides various time-related functions in Python.
import time

# `time.time()` returns the current time in seconds since the epoch (January 1, 1970).
# It gives the number of seconds (as a float) that have passed since this reference time.
# The first `time.time()` call will print the current time at the moment it is called.
print(time.time())

# `time.sleep(2)` pauses the execution of the program for 2 seconds.
# During this time, the program is effectively "sleeping" and doesn't perform any other actions.
time.sleep(2)

# The second `time.time()` call will print the current time again after waiting for 2 seconds.
# The difference between the first and second print statements should be approximately 2 seconds.
print(time.time())
```

```python
# Importing the regular expression module `re` which provides support for working with regular expressions in Python.
import re

# Defining the regular expression pattern to match one or more digits in a string.
# `\d` represents any digit (0-9), and `+` means one or more occurrences.
pattern = r'\d+'

# The text we want to search within. In this case, the string "There are 143 apples".
text = 'There are 143 apples'

# `re.search()` searches the text for the first match of the pattern.
# It returns a match object if found, or None if no match is found.
# In this case, it will find '143' as the first occurrence of one or more digits in the string.
match = re.search(pattern, text)

# If a match is found, `match.group()` will return the part of the string that matches the pattern.
# In this case, it will return the string '143', which is the first match of one or more digits.
# It prints the matched digits ('143') to the console.
print(match.group())
```
