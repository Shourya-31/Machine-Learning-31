```python
to_do_list=["Buy Groceries", "clean the house","Pay bills"]

to_do_list.append("Go for Jog")

to_do_list.remove("clean the house")

print("My To-Do List:",to_do_list)

if("Pay bills" in to_do_list):
    print("You have pending bills to pay.")

print("To do List Remains")
for task in to_do_list:
    print(f"task: {task}")
```

```python
grades = [85, 92, 78, 90, 88]

grades.append(95)

average_grade = sum(grades)/len(grades)

print(f"Average Grade: {average_grade}")

highest_grade = max(grades)
print(f"Highest Grade: {highest_grade}")

lowest_grade = min(grades)
print(f"Lowest Grade: {lowest_grade}")
```

```python
inventory = ["Apples", "Bananas", "Oranges"]

inventory.append("Grades")

inventory.remove("Bananas")

item = "Oranges"

if item in inventory:
    print(f"{item} are in stock")
else:
    print(f"{item} are out of stock")

print("Inventory List: ")

for item in inventory:
    print(f"-{item}")
```

```python
feedback = ["Great Service", "Very satisfied", "Can be better", "Excellent"]

feedback.append("Not happy with service")

# A comment is considered positive if it contains the word "great" or "excellent" (case-insensitive)

positive_feedback_count = sum(                    # sum() is a built-in function that adds up all the numbers in an iterable
    1                                             # We are adding 1 for each comment that matches the condition (so we are counting)
    for comment in feedback                       # This is a generator expression that loops over each 'comment' in the 'feedback' list
    if "great" in comment.lower()                 # comment.lower() converts the comment to lowercase so the check is case-insensitive
       or "excellent" in comment.lower()          # we check if either "great" or "excellent" is present in the comment
)

print("User feedback:")

for comment in feedback:
    print(f"-{comment}")
```
