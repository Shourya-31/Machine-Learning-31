```python
nums=[1,2,3,4,5]

list(map(lambda x:x**2, nums))
```

```python
## Map multiple iterables

nums1=[1,2,3]
nums2=[4,5,6]

added_nums=list(map(lambda x,y:x+y,nums1,nums2))
print(added_nums)
```

```python
## map() to convert list of strings to integers
# Use map to convert strings to integers

str_nums=['1','2','3','4','5']
int_nums=list(map(int,str_nums))

print(int_nums)
```

```python
words=['apple','banana','cherry']
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name':'Krish','age':32},
    {'name':'Sho','age':23}
]

list(map(get_name,people))
```
