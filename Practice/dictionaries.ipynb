```python
## Creating Dictionaries

empty_dict = {}
print(type(empty_dict))
```

```python
student = {"name":"Shourya","age":24,"courses":["Math","CompSci"]}
print(student)
```

```python
## Error cases- Duplicate Keys

student = {"name":"Shourya","age":24, "name":"23", "courses":["Math","CompSci"]}
print(student)
```

```python
## Accessing Values in Dictionaries
student = {"name":"Shourya","age":24, "Grade":'A', "courses":["Math","CompSci"]}
print(student["name"])
print(student["courses"])
print(student["age"])
```

```python
## Accessing using get() method
print(student.get("Grade"))
print(student.get("Lastname"))
print(student.get("Lastname","Not Found"))
```

```python
## Modifying Dictionary Elements
## Dictionaries are mutable but keys are immutable

student["age"] = 31 ## Modifying existing key
student["courses"].append("Statistics")
print(student)
student["Lastname"] = "Veer Singh" ## Adding new key-value pair
print(student)
del student["Grade"]  ## Deleting key-value pair
print(student)
```

```python
## Dictionary Methods

keys = student.keys()
print(keys)
values = student.values()
print(values)

items = student.items()
print(items)
```

```python
## shallow copy of dictionary

student1 = student.copy()
print(student1)
```

```python
student["name"] = "Yo Sho"

print(student1)
```

```python
student2 = student.copy()
print(student1)
print(student2)
```

```python
## Iterating through a dictionary
## Can use loops to iterate through keys, values or key-value pairs

## Iterating through keys
for key in student.keys():
    print(key)
```

```python
## Iterating through values
for value in student.values():
    print(value)
```

```python
## Iterating through key-value pairs
for key, value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries

heroes = {
    "hero1":{
        "name":"Spiderman",
        "age":26
    },
    "hero2":{
        "name":"Ironman",
        "age":48
    }
}

print(heroes)
```

```python
## Accessing Nested Dictionaries Elements
print(heroes["hero1"]["name"])
```

```python
student.items()
```

```python
## Iterating over nested dictionaries

for student_id,student_info in student.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehensions

sqaures = {x:x**2 for x in range(5)}
print(sqaures)
```

```python
## Conditional Dictionary Comprehensions

sqrs = {x:x**2 for x in range(10) if x%2==0}
print(sqrs)
```

```python
## Ex: Count frequency of elements in a list

nums = [1,2,2,3,4,4,4,5,5,5,5]

freq = {}

for num in nums:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1
print(freq)
```

```python
## Merge Dictionaries

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
```
