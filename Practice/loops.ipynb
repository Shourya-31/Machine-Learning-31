```python
## while loop

count = 0

while count<5:
    print("Count is: ", count)
    count += 1
```

```python
count  = 0

while count%2 == 0:
    print(count)
    count=count+1
```

```python
## Loop control statements

## break statement exits the loop prematurely

for i in range(10):
    if i==9:
        break
    else:
        print(i)
```

```python
## continue statement

## skips the current iteration and moves to the next iteration

for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
## pass
## is a null statement, does nothing, used as a placeholder code

for i in range (5):
    if i == 3:
        pass
    print(i)
```

```python
## Nested loops

for i in range (5):
    for j in range(3):
        print(f"i:{i} and j: {j}")
```

```python
## Examples- Calculate the sum of first N natural numbers using a while and for loop

n = int(input("Enter a Number: "))

sum = 0
count = 1

while count<=n:
    sum+=count
    count+=1

print(f"Sum of first {n} natural numbers is: {sum}")
```

```python
## Example- Prime numbers between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
