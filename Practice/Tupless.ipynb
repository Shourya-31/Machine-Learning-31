```python
## Immutable nature of Tuples
## They cannot be changed once created

lst1 = [1,2,3]
tpl1 = (1,2,3)

lst1[0]="Sho"
print(lst1)
## tpl1[0]="Sho" ## TypeError: 'tuple' object does not support item assignment
print(tpl1)
```

```python
## Tuple Methods

print(nums.count(3))
print(nums.index(4))
```

```python
## Packing and Unpacking Tuples

packed_tpl = 1,"Hiiii",3.14
print(packed_tpl)
```

```python
## Unpacking Tuples

a,b,c = packed_tpl
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
## * makes a list of remaining values till the last element

numm=(1,2,3,4,5,6,7,8,9)
first,*middle,last=numm
print(first)
print(middle)
print(last)
```

```python
## Nested Tuples

lst = [[1,2,3],[4,5,6],[7,8,9]]
lst[1][0:2]
```

```python
lst = [[1,2,3],[4,5,6],[7,8,9]]
lst[0][2]
```

```python
lst = [[1,2,3],[4,5,6],[7,8,9]]

print(lst[0][2])
print(lst[1][0:2])
```

```python
## Iterating through Nested Tuples

for sub_tpl in lst:
    for item in sub_tpl:
        print(item,end=" ")
    print()
```
