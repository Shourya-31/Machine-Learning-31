```python
## Basic Magic Methods
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def __str__(self):
        return f"{self.name},{self.age} years old"
person=Person("SHo",23)
print(person)
```

```python
## Basic Magic Methods
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def __str__(self):
        return f"{self.name},{self.age} years old"
    
    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
person=Person("Sho",23)
print(person)
```
