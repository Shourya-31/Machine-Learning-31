```python
class Cat:
    #constructor
    def __init__(self,meowAge,meowName):
        self.meowAge=meowAge
        self.meowName=meowName
    
    def meow(self):
        print(f"{self.meowName} says Meowwwww~")


c1=Cat(2,"Meu")

c1.meow()
print(c1.meowAge)
print(c1.meowName)
    
```

```python
## Modeling a Bank Account

## Define a class for bank account:
class BankAcc:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} is deposited. New balance is: {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient Funds")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn. New balance is: {self.balance}")

    def get_balance(self):
        return self.balance
    
## Create an object:

account=BankAcc("Shourya",1000)
print(account.balance)

account.withdraw(100)
account.deposit(500)
print(account.get_balance())

```

```python
class V:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self):
        print(f"The person will drive the {self.enginetype} car")

c1 = V(4, 5, "Diesel")
c1.drive()
```

```python
class Tesla(V):
    def __init__(self, windows, doors, enginetype,is_selfDriving):
        super().__init__(windows, doors, enginetype) 
        self.is_selfDriving=is_selfDriving
    
    def selfDrive(self):
        print(f"The car is self driving: {self.is_selfDriving}")

t1=Tesla(4,5,"Electric",True)
t1.selfDrive()
```

```python
## Multiple Inheritance
class Animal:
    def __init__(self,name):
        self.name=name
    
    def speak(self):
        print("SUbclass must implement this method")

class Pet:
    def __init__(self,owner):
        self.owner=owner

class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} says woof"
    
dog=Dog("Buddy","Sho")
print(dog.speak())
```
