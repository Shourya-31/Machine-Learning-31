```python
## Simple way to create iterators
## Uses yield keyword to produce a series of values lazily
## They generate values on fly and dont store them in memory
```

```python
def square(n):
    for i in range(3):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a=square(3)
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
    
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

```python
## Practical example: Reading Large Files
## Processes file line by line without loading into memory

def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
```

```python
file_path='largee.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
