```python
## Iterators:

my_list=[1,2,3,4,5]
iterator=iter(my_list)
print(type(iterator))
```

```python
my_list=[1,2,3,4,5]
iterator=iter(my_list)

try:
    print(next(iterator))
except StopIteration:
    print("No elements left in list")
```
