```python
lst = len([1, 2, 3, 4, 5])

def main_welcome(func, lst):
    msg = "Welcome"
    def sub_welcome():
        print("Welcome hereee!")
        print(func(lst))
        print("Have a good day")

    return sub_welcome  # return the function, don't call it here

# Pass a function that returns a value (e.g., returning the length of the list)
def print_length(lst):
    return f"The length of the list is: {lst}"

# Now we can call the function
main_welcome(print_length, lst)()  # We call it to execute the sub_welcome function
```

```python
## Decorators

def main_welcome(func):
    msg = "Welcome"
    def sub_welcome():
        print("Welcome hereee!")
        func()
        print("Have a good day")

    return sub_welcome()

@main_welcome
def course_intro():
    print("This is an advanced python course")
```

```python
## Decorators with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hii():
    print("Hiii")

say_hii()
```

```python
# Define the outer decorator function `repeat` which takes an argument `n`
def repeat(n):
    # This is the actual decorator function that takes a function `func` as an argument
    def decorator(func):
        # This is the wrapper function which will replace the original function
        def wrapper(*args, **kwargs):
            # This loop will run the decorated function `n` times
            for _ in range(n):
                # Call the original function each time
                func(*args, **kwargs)
        # Return the wrapper function which is now the decorated version of `func`
        return wrapper
    # Return the decorator itself (which will be used to decorate a function)
    return decorator

# Now use the decorator by applying `@repeat(3)` on the `say_hii` function
@repeat(3)
def say_hii():
    print("Hiii")

# Calling `say_hii()` will actually call the wrapper, which will execute the print statement 3 times
say_hii()
```
