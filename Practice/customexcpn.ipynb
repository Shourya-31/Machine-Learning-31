```python
## Custom Exception Class

class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
class dobException(Exception):
    """Custom exception for invalid age"""
    pass

year = int(input("Enter the DOB: "))
age = 2024 - year

try:
    if 20 <= age <= 30:
        print("Age is valid")
    else:
        raise dobException
except dobException:
    print("Sorry, age is not valid")
```
