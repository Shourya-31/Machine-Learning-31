```python
## Read a Whole File

with open('ex11.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## Read a file line by line

with open('ex11.txt','r') as file:
    for line in file:
        print(line.strip()) ## strip() removes the newline character
```

```python
## Writing a file (overwriting)

with open('ex11.txt','w') as file:
    file.write('Hello there new world!\n')
    file.write('I am Shourya')
```

```python
## Write a file (without overwriting)

with open('ex11.txt','a') as file:
    file.write("Its appending \n")
```

```python
## Writing a list of lines to a file

lines=['\n','First line \n','Second line \n','Third line \n']
with open('ex11.txt','a') as file:
    file.writelines(lines)
```

```python
## Binary files

# Writing to a binary file
data=b'\x00\x01\x02\x03\x04'
with open('ex.bin','wb') as file: # wb=writeByte mode
    file.write(data)
```

```python
# Reading a binary file
with open('ex.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
## Read the content from a source text file and write it to a destination text file
#copying a text file
with open('ex11.txt','r') as source_file:
    content = source_file.read()

with open('destinations1.txt','w') as detination_file:
    detination_file.write(content)
```
