```python
import pandas as pd

df = pd.read_csv('dataz.csv')

df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)

# Preview the DataFrame
df.head()
```

```python
df['New Value']=df['Value'].apply(lambda x:x**2)
df.head()
```

```python
# Grouping the DataFrame 'df' by the 'Product' column
# This creates groups based on unique values in the 'Product' column
grouped_mean = df.groupby('Product')['Value'].mean()

# The above line calculates the mean (average) of the 'Value' column for each group of 'Product'
# Essentially, it's aggregating the data by 'Product' and computing the average of 'Value' for each product

# Printing the result which shows the mean of 'Value' for each unique 'Product'
print(grouped_mean)
```

```python
df.groupby(['Product','Region'])['Value'].sum()
```

```python
df.groupby(['Product','Region'])['Value'].mean()
```

```python
## Aggregate Multiple functions
grouped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
grouped_agg
```

```python
## Merging and joining Dataframes
df1=pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2=pd.DataFrame({'Key':['A','B','D'], 'Value2':[4,5,6]})

df2
```

```python
## Merge Dataframe on the 'Key Coloumns'
pd.merge(df1,df2,on="Key",how="inner")
# on=which coloumn, and based on matching between Keys, inner joins happens
```

```python
pd.merge(df1,df2,on="Key",how="outer")
# for "outer", for the dataframe will be considered
#since Value2 doesnt have value for C so NAN and same goes for D in Value1
```

```python
pd.merge(df1,df2,on="Key",how="left")
# left outer join => left DF will be given more priority (df1)
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
