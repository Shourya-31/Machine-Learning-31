```python
## Series
## A Pandas Series is a one-dimensional array-like object that
## can hold any data type. It is similar to a column in a table.
```

```python
import pandas as pd
data=[1,2,3,4,5,6]
series=pd.Series(data)

print("Series:\n",series)
```

```python
## Create a series from dictionary
data={'a':1,'b':2,'c':3}
dict_series=pd.Series(data)
print(dict_series)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
## Create a Dataframe from a dictionary of list

data={
    'Name':['Sho','Yo','Wo'],
    'Age':[23,56,78],
    'City':['Mumbai','Bilaspur','Raipur']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a Data from a list of dictionaries

data=[
    {'Name':'Sho','Age':23,'City':'Mumbai'},
    {'Name':'Sho1','Age':22,'City':'Mumbai1'},
    {'Name':'Sho2','Age':24,'City':'Mumbai2'},
    {'Name':'Sho3','Age':25,'City':'Mumbai3'}
]

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'Sales': [20000, 25000, 30000]
})

# Add a new column
df['Profit'] = [5000, 6000, 7000]

print("Original DataFrame:")
print(df)

# Remove a column
df.drop('Profit', axis=1, inplace=True)

print("\nDataFrame after removing 'Profit':")
print(df)
```
