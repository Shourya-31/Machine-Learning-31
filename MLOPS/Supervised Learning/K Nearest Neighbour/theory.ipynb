## CLassification

#### Lets say we have binary categories

#### First Step- We need to establish a K value

#### This can change based on various datasets, that implies its a hyperparameter

#### Second Step- Find K nearest neighbour for the test data

#### Third Step- From those K = X how many neighbour belongs to 0 category

#### Now for this we use 2 different kind of formulas:-

####  1.) Euclidian Distance 
####  2.) Manhatton Distance


### Second Classification:- Regression

#### Here we basically find the average of 5 nearest neighbours and get the o/p like price of house is avg of 5 nearest neighbours

### We optimize the calculation with help of:-

### 1.) KD Tree
### 2.) Binary Tree

Our main aim is that whenever we find the k nearest neighbour we shouldnt check the distance between every point.

Therefore we make a KD Tree which is in format of a Binary Tree

We are reducing number of searches

We basically divided the region based on the points and medians thus making a binary tree AKA KD Tree


In short we are making groups such that if a new query comes, we will know it belongs to this region

This helps om decreasing the Time Complexity well
