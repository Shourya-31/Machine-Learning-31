# Decision Tree Classifier

A **Decision Tree Classifier** is a supervised machine learning algorithm used for **classification problems**.
It works like a flowchart where:

* Each **internal node** represents a condition/test on a feature.
* Each **branch** represents the outcome of the test.
* Each **leaf node** represents the final class prediction.

## Working

The algorithm repeatedly splits the dataset into smaller subsets based on the feature that gives the **best separation** between classes.

Common splitting criteria:

* **Gini Impurity**
* **Entropy / Information Gain**

## Steps

1. Select the best feature for splitting.
2. Divide the dataset into branches.
3. Repeat splitting recursively.
4. Stop when data is pure or stopping condition is reached.

## Advantages

* Easy to understand and visualize.
* Works with numerical and categorical data.
* Requires little data preprocessing.
* Handles non-linear relationships well.

## Disadvantages

* Can overfit on training data.
* Sensitive to small changes in data.
* Large trees become complex.

## Applications

* Spam detection
* Medical diagnosis
* Customer segmentation
* Fraud detection

## Example

If predicting whether a person buys a laptop:

* Age < 30?

  * Yes → Student?

    * Yes → Buy
    * No → Not Buy
  * No → Income High?

    * Yes → Buy
    * No → Not Buy

## Important Terms

* **Root Node**: Starting node of the tree.
* **Leaf Node**: Final prediction/output node.
* **Depth**: Maximum level of the tree.
* **Pruning**: Removing unnecessary branches to reduce overfitting.
