```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = 6*np.random.rand(100,1) - 3 ## Outlier np.random.rand(100,1)
y = 0.5*x**2 + 1.5*x + 2 + np.random.rand(100,1)
plt.scatter(x,y,color='g')
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
plt.show()
```

### Implies Polynomial Regression


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
## Lets implement a simple LR
from sklearn.linear_model import LinearRegression
regression1=LinearRegression()
```

```python
regression1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,regression1.predict(X_test))
print(score)
```

```python
## Lets visualize this model
plt.plot(X_train,regression1.predict(X_train),color='r')
plt.scatter(X_train,y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y dataset")
plt.show()
```

## Now implementing Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.fit_transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
regression = LinearRegression()
regression.fit(X_train_poly, y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
plt.scatter(X_train,regression.predict(X_train_poly))
plt.show()
```

```python
plt.scatter(X_train,y_train)
plt.show()
```

```python
poly=PolynomialFeatures(degree=3,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.fit_transform(X_test)
```

```python
X_train_poly
```

```python
regression = LinearRegression()
regression.fit(X_train_poly, y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

### Degree = 2 was working finer

```python
## Prediction of new dataset
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_poly=poly.transform(X_new)
```

```python
X_new_poly
```

```python
y_new=regression.predict(X_new_poly)
plt.plot(X_new,y_new,"r-",linewidth=2,label="Predictions")
plt.plot(X_train,y_train,"b.",label="Training points")
plt.plot(X_test,y_test,"g.",label='Testing Points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
