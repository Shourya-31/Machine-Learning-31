```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = 6*np.random.rand(100,1) - 3 ## Outlier np.random.rand(100,1)
y = 0.5*x**2 + 1.5*x + 2 + np.random.rand(100,1)
plt.scatter(x,y,color='g')
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
plt.show()
```

### Implies Polynomial Regression


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
## Lets implement a simple LR
from sklearn.linear_model import LinearRegression
regression1=LinearRegression()
```

```python
regression1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,regression1.predict(X_test))
print(score)
```

```python
## Lets visualize this model
plt.plot(X_train,regression1.predict(X_train),color='r')
plt.scatter(X_train,y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y dataset")
plt.show()
```

## Concept of Pipelines

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    # Generate 200 evenly spaced points between -3 and 3.
    # These will be used to visualize the regression curve smoothly.
    X_new = np.linspace(-3, 3, 200).reshape(200, 1)

    # Create polynomial features of the given degree.
    # include_bias=True adds a bias (x^0 = 1) term to the data.
    poly_features = PolynomialFeatures(degree=degree, include_bias=True)

    # Initialize a simple Linear Regression model.
    line_reg = LinearRegression()

    # Create a pipeline that first transforms X into polynomial features,
    # then fits a linear regression model on those features.
    # This avoids manually transforming and then fitting separately.
    poly_regression = Pipeline([
        ("poly_features", poly_features),
        ("line_reg", line_reg)
    ])

    # Fit the model on training data (X_train, y_train).
    # The model will learn the polynomial relationship between X and y.
    poly_regression.fit(X_train, y_train)

    # Predict the y-values for our new X_new data (the smooth line points).
    y_pred_new = poly_regression.predict(X_new)

    
    # --- Plotting ---
    # Plot the polynomial regression prediction as a red line ('r').
    plt.plot(X_new, y_pred_new, 'r', label="Degree " + str(degree), linewidth=2)

    # Plot the original training data as blue dots ('b.').
    plt.plot(X_train, y_train, "b.", linewidth=3)

    # Add legend to identify which polynomial degree the red line represents.
    plt.legend(loc="upper left")

    # Label x and y axes.
    plt.xlabel("X")
    plt.ylabel("y")

    # Set axis limits for better visibility of data and curve.
    plt.axis([-4, 4, 0, 10])

    # Display the plot.
    plt.show()


# Example usage:
# Fits and visualizes a 6th-degree polynomial regression curve
poly_regression(2)
```
