- Reduce the loss value. :contentReference[oaicite:6]{index=6}

## Need for Backward Propagation
If loss is high:
- Weights must be updated.

Backward propagation updates weights from:
- Output layer → Hidden layer → Earlier layers

This helps reduce prediction error. :contentReference[oaicite:8]{index=8}

---

## Weight Updation Formula

The main formula used is:

\[
W_{new} = W_{old} - \eta \frac{d(Loss)}{d(W_{old})}
\]

Where:
- \(W_{new}\) = Updated weight
- \(W_{old}\) = Previous weight
- \(\eta\) = Learning rate
- \(\frac{d(Loss)}{d(W_{old})}\) = Derivative of loss w.r.t weight

This is the core formula of gradient descent. :contentReference[oaicite:9]{index=9}

---

## Meaning of Derivative
Derivative represents:
- Slope of the loss curve

It tells:
- Whether to increase weight
- Or decrease weight

This helps move toward minimum loss. :contentReference[oaicite:10]{index=10}

---

## Gradient Descent
Gradient descent is an optimizer used to reduce loss.

Main objective:
- Reach Global Minima

Global minima is the point where:
- Loss becomes minimum
- Slope becomes zero

At this point:
- Further weight updation is unnecessary. :contentReference[oaicite:11]{index=11}

---

## Understanding Slope Direction

### Negative Slope
If the graph points downward:
- Slope is negative
- Weight value increases

Result:
\[
W_{new} > W_{old}
\]

This moves weights toward global minima. :contentReference[oaicite:12]{index=12}

---

### Positive Slope
If the graph points upward:
- Slope is positive
- Weight value decreases

Result:
\[
W_{new} < W_{old}
\]

Again, this helps move toward global minima. :contentReference[oaicite:13]{index=13}

## When Weight Stops Updating
At global minima:
- Slope becomes zero

So:

\[
\frac{d(Loss)}{d(W)} = 0
\]

Thus:

\[
W_{new} = W_{old}
\]

This means:
- Training has converged. :contentReference[oaicite:16]{index=16}
