# Deep Learning: Multilayer Perceptron, Forward Propagation, and Backward Propagation

## Introduction
Single-layer perceptrons have limitations, so we use multilayer perceptrons, also called artificial neural networks. The lecture focuses on forward propagation, backward propagation, loss functions, optimizers, and activation functions. :contentReference[oaicite:1]{index=1}

## What is Multilayer Perceptron?
A multilayer perceptron is a multilayered neural network. It contains:
- Input layer
- One or more hidden layers
- Output layer

This structure allows the network to learn more complex patterns than a single-layer perceptron. :contentReference[oaicite:2]{index=2}

## Why Multilayer Networks are Needed
A single-layer perceptron only performs feedforward computation and is limited in solving complex problems. Multilayer networks solve this by using forward propagation and backward propagation to train weights efficiently. :contentReference[oaicite:3]{index=3}

## Geoffrey Hinton and Backpropagation
Backward propagation is closely associated with Geoffrey Hinton, who is credited for major contributions to deep learning. His work made it possible to train deep neural networks effectively. :contentReference[oaicite:4]{index=4}

The lecture uses a sample dataset with:
- IQ

The output is:

## Neural Network Structure
The network contains:
- Input layer
- Hidden layer 1
- Hidden layer 2
- Output layer

Each layer has neurons, and each neuron uses weights and bias values to compute its output. :contentReference[oaicite:6]{index=6}

## Weights and Bias
Weights connect neurons between layers. Bias is an additional constant added to the weighted sum. Both weights and biases are initialized randomly at the beginning of training. :contentReference[oaicite:7]{index=7}

## Forward Propagation
Forward propagation means passing input from the input layer to the output layer step by step.

It has two main parts:
1. Weighted summation
2. Activation function

This process happens in every neuron of the network. :contentReference[oaicite:8]{index=8}

## Step 1: Weighted Sum
The neuron first computes:

z = x1w1 + x2w2 + x3w3 + b

This is the sum of inputs multiplied by weights, plus bias. :contentReference[oaicite:9]{index=9}

## Step 2: Activation Function
After computing z, the neuron applies an activation function. In the lecture, sigmoid is used as the main example. :contentReference[oaicite:10]{index=10}

## Sigmoid Activation Function
Sigmoid maps any input value into the range 0 to 1.

Formula:

sigmoid(z) = 1 / (1 + e^-z)

This helps the neuron decide whether to activate or not. :contentReference[oaicite:11]{index=11}

## Example Calculation for Hidden Layer 1
Given:
- x1 = 95
- x2 = 4
- x3 = 4
- w1 = 0.01
- w2 = 0.02
- w3 = 0.03
- b1 = 0.001

Then:

z = (95 × 0.01) + (4 × 0.02) + (4 × 0.03) + 0.001  
z = 1.151

After applying sigmoid:

o1 = sigmoid(1.151) = 0.759

So the output of hidden layer 1 is 0.759. :contentReference[oaicite:12]{index=12}

## Example Calculation for Hidden Layer 2
Given:
- o1 = 0.759
- w4 = 0.02
- b2 = 0.03

Then:

z = (0.759 × 0.02) + 0.03  
z = 0.04518

After applying sigmoid:

o2 = sigmoid(0.04518) = 0.51129

So the final predicted output becomes 0.51129. :contentReference[oaicite:13]{index=13}

## Loss Function
The loss function measures the difference between the predicted output and the actual output.

If:
- Actual output = 1
- Predicted output = 0.51129

Then the error is high, and the model must reduce it during training. The lecture describes this as the main purpose of the loss function. :contentReference[oaicite:14]{index=14}

## Goal of Training
The main goal is to minimize loss by updating weights properly. This is what allows the neural network to learn from data. :contentReference[oaicite:15]{index=15}

## Backward Propagation
Backward propagation updates the weights starting from the output layer and moving backward toward the hidden layers.

The lecture explains that:
- First, w4 is updated
- Then w1, w2, and w3 are updated

This process reduces the loss and improves model performance. :contentReference[oaicite:16]{index=16}

## Optimizers
Optimizers are used to reduce the loss function and update weights efficiently during backward propagation. They make training faster and more effective. :contentReference[oaicite:17]{index=17}

## Loss Function vs Cost Function
Loss function is calculated for a single example.

Cost function is calculated over the entire dataset. It represents the overall training error across all samples. :contentReference[oaicite:18]{index=18}

## Complete Workflow
1. Initialize weights and biases
2. Perform forward propagation
3. Compute output
4. Calculate loss
5. Perform backward propagation
6. Update weights using optimizer
7. Repeat until loss becomes small

## Key Takeaway
A multilayer neural network learns by repeatedly doing forward propagation and backward propagation. Forward propagation produces predictions, loss function measures the error, and backward propagation updates weights to reduce that error. :contentReference[oaicite:19]{index=19}
