#### In Deep Learning we will be working with Multi-layered Neural Networks

## Perceptron

#### It is a single layered Neural Network. This is the basic unit of ANN AKA Artificial Neuron.

#### This is used in solving Binary Classification.

#### Perceptron has input layer. A hidden Layer which has just one neuron. We do not count i/p layer as seperate layer. It has weights to decide the influence of inputs. Now in Hidden layer we have a hidden neuron and inside the neuron we have 2 sinmple steps which will happen. In step-1 we will do summation of all inputs and weights. The i/p are never restricted. So they can be of any number. 


#### With this we will have a bias b. 
#### Together this summation plus bias will make Z as step-1.

#### This bias needs to have a value so o.p doesnt become zero, this bias is treated as noise. In linear regression term we call it as Intercept. 


#### Now in step-2, we will assign an activation function. The Activation function transforms the output between 0 to 1 and -1 to 1. It is in form of step function, sigmoid function, relu, etc.

#### To solve any binary classifcation problem we apply activation function.
