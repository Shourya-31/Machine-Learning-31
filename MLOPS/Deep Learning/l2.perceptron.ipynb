#### In Perceptron we have 2 types of perceptrons, first single layered perceptron and then comes multi-layered perceptron.

#### Single layered process is also called, "Feed Forward Neural Network"

#### Linear Seperable Problems can be easily seperated via single layered perceptron. 

#### For Non-Linear seperable Problems, it cant be solved just by single layered perceptron. Therefore we need Multi-Layered Neural Network for it.
